**1) Print customer Id, customer name and average account_balance maintained by each customer for all  of his/her accounts in the bank.(8 Rows)**

```python
cursor.execute('''
SELECT
    b.Customer_id,
    a.customer_name,
    AVG(b.Balance_amount) AS avg_balance
FROM
   BANK_ACCOUNT_DETAILS b
JOIN
   BANK_CUSTOMER a
ON
   b.Customer_id = a.customer_id
GROUP BY
    b.Customer_id,
    a.customer_name;
''')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

**2) Print customer_id , account_number and balance_amount , condition that if balance_amount is nil then assign transaction_amount  for account_type = "Credit Card"(4 Rows)**

```python
cursor.execute('''
SELECT
    b.Customer_id,
    b.Account_Number,
    CASE
        WHEN b.Balance_amount IS NULL THEN t.Transaction_amount
        ELSE b.Balance_amount
    END AS final_amount
FROM
    BANK_ACCOUNT_DETAILS b
LEFT JOIN
    BANK_ACCOUNT_TRANSACTION t
ON
    b.Account_Number = t.Account_Number
WHERE
    b.Account_type = 'Credit Card'
LIMIT 4;
''')

rows = cursor.fetchall()

for row in rows:
  print(row)
```

**3) Print account_number and balance_amount , transaction_amount,Transaction_Date from Bank_Account_Details and bank_account_transaction for all the transactions occurred during march,2020 and april, 2020(12 Rows)**

```python
cursor.execute('''
SELECT
    b.Account_Number,
    b.Balance_amount,
    t.Transaction_amount,
    t.Transaction_Date
FROM BANK_ACCOUNT_DETAILS b
JOIN BANK_ACCOUNT_TRANSACTION t
    ON b.Account_Number = t.Account_Number
WHERE
    strftime('%Y-%m', t.Transaction_Date) IN ('2020-03', '2020-04')
LIMIT 12;
''')
rows = cursor.fetchall()

for row in rows:
  print(row)
```

**4) Print all of the customer id, account number,  balance_amount, transaction_amount , Transaction_Date  from bank_customer, Bank_Account_Details and bank_account_transaction tables where excluding all of their transactions in march, 2020  month (22 Rows)**

```python
cursor.execute('''
SELECT
    b.Customer_id,
    a.Account_Number,
    a.Balance_amount,
    t.Transaction_amount,
    t.Transaction_Date
FROM
    BANK_CUSTOMER b
JOIN
    BANK_ACCOUNT_DETAILS a
ON
    b.Customer_id = a.Customer_id
JOIN
    BANK_ACCOUNT_TRANSACTION t
ON
    a.Account_Number = t.Account_Number
WHERE
     strftime('%Y-%m', t.Transaction_Date) != '2020-03';
'''
)
rows = cursor.fetchall()

for row in rows:
  print(row)
```

**7)Print all Customer_id, Account_Number, Account_type, and display deducted balance_amount by  subtracting only negative transaction_amounts for Relationship_type ="P" ( P - means  Primary , S - means Secondary ) .(27 Rows)**

```python
cursor.execute('''
SELECT
    b.Customer_id,
    b.Account_Number,
    b.Account_Type,
    b.Balance_amount -
        CASE
            WHEN negative_tx.total_negative IS NULL THEN 0
            ELSE negative_tx.total_negative
        END AS deducted_balance
FROM BANK_ACCOUNT_DETAILS b
LEFT JOIN (
    SELECT
        Account_Number,
        SUM(
            CASE
                WHEN Transaction_amount < 0 THEN Transaction_amount
                ELSE 0
            END
        ) AS total_negative
    FROM BANK_ACCOUNT_TRANSACTION
    GROUP BY Account_Number
) negative_tx
    ON b.Account_Number = negative_tx.Account_Number
WHERE b.Relationship_type = 'P';
''')
rows = cursor.fetchall()

for row in rows:
  print(row)
```

**10)Display all saving account holders have Add-on Credit Cards" and Credit cards" (3 Rows)**

```python
cursor.execute('''
SELECT DISTINCT
    s.Customer_id,
    s.Account_Number AS Saving_Account,
    c.Account_Number AS Credit_Card_Account
FROM BANK_ACCOUNT_DETAILS s
JOIN BANK_ACCOUNT_DETAILS c
    ON s.Customer_id = c.Customer_id
WHERE s.Account_Type = 'Saving'
  AND c.Account_Type IN ('Credit Card', 'Add-on Credit Card');
''')
rows = cursor.fetchall()

for row in rows:
  print(row)
```

**11)  Display  records of SAVINGS account linked with Credit cards" account_type and its credit**

```python
cursor.execute("""
SELECT
    s.account_number AS savings_account,
    c.account_number AS credit_card_account,
    SUM(t.transaction_amount) AS total_credit_transactions
FROM
    Bank_Account_Details s
JOIN
    Bank_Account_Relationship_Details r
    ON s.account_number = r.account_number
JOIN
    Bank_Account_Details c
    ON r.linking_account_number = c.account_number
JOIN
    BANK_ACCOUNT_TRANSACTION t
    ON c.account_number = t.account_number
WHERE
    s.account_type = 'SAVINGS'
    AND c.account_type = 'Credit Card';
""")
rows = cursor.fetchall()

for row in rows:
  print(row)
```
